```python
import os
from moviepy import VideoFileClip


def analyze_videos(folder_path):
  files = [file for file in os.listdir(folder_path) if file.endswith(".mp4")]
  total_files = len(files)
  total_size = sum(os.path.getsize(os.path.join(folder_path, file)) for file in files)

  print(f"Total video files to process: {total_files}")
  print(f"Total directory size: {total_size / (1024 * 1024):.2f} MB\n")

  processed_files = 0
  total_duration = 0

  for i, file in enumerate(files, start=1):
    print(f"#{i} Processing: {file}")
    filepath = os.path.join(folder_path, file)
    processed_files += 1
    clip = VideoFileClip(filepath)
    total_duration += clip.duration
    clip.close()

  print(f"\nTotal processed files: {processed_files}")
  print(f"Total size: {total_size / (1024 * 1024):.2f} MB")
  print(f"Total duration: {total_duration / 60:.2f} minutes")


folder = r"C:\Users\k26ra\Videos\Captures"
analyze_videos(folder)
```

```python
import os
from moviepy import VideoFileClip


def compare_videos(folder_path):
  mkv_files = [file for file in os.listdir(folder_path) if file.endswith(".mkv")]
  mp4_files = {file for file in os.listdir(folder_path) if file.endswith(".mp4")}

  unmatched_files = []

  for mkv_file in mkv_files:
    mp4_file = os.path.splitext(mkv_file)[0] + ".mp4"
    if mp4_file in mp4_files:
      mkv_path = os.path.join(folder_path, mkv_file)
      mp4_path = os.path.join(folder_path, mp4_file)

      with VideoFileClip(mkv_path) as mkv_clip, VideoFileClip(mp4_path) as mp4_clip:
        if abs(mkv_clip.duration - mp4_clip.duration) > 1:  # Allow slight difference
          unmatched_files.append(mkv_file)
    else:
      unmatched_files.append(mkv_file)

  if unmatched_files:
    print("Unmatched MKV files:")
    for file in unmatched_files:
      print(file)
  else:
    print("All MKV files have matching MP4 files with the same duration.")


folder = r"C:\Users\k26ra\Videos"
compare_videos(folder)
```

```python
import os
from moviepy import VideoFileClip


def compare_videos(folder_path):
  mkv_files = [file for file in os.listdir(folder_path) if file.endswith(".mkv")]
  mp4_files = {file for file in os.listdir(folder_path) if file.endswith(".mp4")}

  all_mp4_found = True

  for mkv_file in mkv_files:
    mp4_file = os.path.splitext(mkv_file)[0] + ".mp4"
    mkv_path = os.path.join(folder_path, mkv_file)
    mkv_size = os.path.getsize(mkv_path) / (1024 * 1024)

    try:
      with VideoFileClip(mkv_path) as mkv_clip:
        mkv_duration = f"{mkv_clip.duration:.2f} sec"
    except Exception:
      mkv_duration = "NA"

    print(f"MKV: {mkv_file} | Size: {mkv_size:.2f} MB | Duration: {mkv_duration}")

    if mp4_file in mp4_files:
      mp4_path = os.path.join(folder_path, mp4_file)
      mp4_size = os.path.getsize(mp4_path) / (1024 * 1024)

      try:
        with VideoFileClip(mp4_path) as mp4_clip:
          mp4_duration = f"{mp4_clip.duration:.2f} sec"
      except Exception:
        mp4_duration = "NA"

      print(f"MP4: {mp4_file} | Size: {mp4_size:.2f} MB | Duration: {mp4_duration}")

      if mkv_duration == "NA" or mp4_duration == "NA" or mkv_duration == mp4_duration:
        print("✅ MP4 found and durations match or NA.")
      else:
        print("❌ MP4 found but durations do not match.")
    else:
      print("❌ No matching MP4 file found.")
      all_mp4_found = False

  print("\n========================================")
  if all_mp4_found:
    print("🎉🎉🎉 MISSION SUCCESSFUL: ALL MKV FILES HAVE MATCHING MP4 FILES! 🎉🎉🎉")
  else:
    print("🚨🚨🚨 MISSION FAILED: SOME MKV FILES ARE MISSING MATCHING MP4 FILES! 🚨🚨🚨")


folder = r"C:\Users\k26ra\Videos"
compare_videos(folder)
```

```python
import os
from moviepy import VideoFileClip


def get_video_details(file_path):
  """Get the size and duration of a video file."""
  size = os.path.getsize(file_path) / (1024 * 1024)  # Size in MB
  try:
    with VideoFileClip(file_path) as clip:
      duration = f"{clip.duration:.2f} sec"
  except Exception:
    duration = "NA"
  return size, duration


def print_video_details(label, file_name, size, duration):
  """Print video details in a clean format."""
  print(f"{label}: {file_name} | Size: {size:.2f} MB | Duration: {duration}")


def check_mkv_mp4_pair(mkv_path, mp4_path):
  """Check if MKV and MP4 pair match and print their details."""
  mkv_size, mkv_duration = get_video_details(mkv_path)
  print_video_details("MKV", os.path.basename(mkv_path), mkv_size, mkv_duration)

  if mp4_path:
    mp4_size, mp4_duration = get_video_details(mp4_path)
    print_video_details("MP4", os.path.basename(mp4_path), mp4_size, mp4_duration)

    if mkv_duration == "NA" or mp4_duration == "NA" or mkv_duration == mp4_duration:
      print("✅ MP4 found and durations match or NA.")
    else:
      print("❌ MP4 found but durations do not match.")
  else:
    print("❌ No matching MP4 file found.")


def compare_videos(folder_path):
  """Main function to compare MKV and MP4 files in a folder."""
  mkv_files = [file for file in os.listdir(folder_path) if file.endswith(".mkv")]
  mp4_files = {file: os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(".mp4")}

  all_mp4_found = True

  for mkv_file in mkv_files:
    mkv_path = os.path.join(folder_path, mkv_file)
    mp4_path = mp4_files.get(os.path.splitext(mkv_file)[0] + ".mp4")

    check_mkv_mp4_pair(mkv_path, mp4_path)

    if not mp4_path:
      all_mp4_found = False

  print("\n========================================")
  if all_mp4_found:
    print("🎉🎉🎉 MISSION SUCCESSFUL: ALL MKV FILES HAVE MATCHING MP4 FILES! 🎉🎉🎉")
  else:
    print("🚨🚨🚨 MISSION FAILED: SOME MKV FILES ARE MISSING MATCHING MP4 FILES! 🚨🚨🚨")


folder = r"C:\Users\k26ra\Videos"
compare_videos(folder)
```
